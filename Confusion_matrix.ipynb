```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Confusion matrix RoBERTa statement
```

```python
# Create a random confusion matrix with 6 labels
np.random.seed(45)  
num_classes = 6
conf_matrix = np.array([
    [117, 141, 3,  19,  10,  13],  # Pants on Fire
    [86,  389, 18, 67,  49,  51],  # False
    [9,   120, 23, 107, 58,  43],  # Mostly False
    [4,   82,  16, 125, 84,  60],  # Half True
    [1,   90,  5,  71, 107, 69],   # Mostly True
    [1,   62,  2,  37, 79,  78]    # True
])

# Class labels
class_labels = ["Pants on Fire", "False", "Mostly False", "Half True", "Mostly True", "True"]

# Create a heatmap for the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.title("Confusion Matrix RoBERTa with Statement Feature", fontsize=12, fontweight='bold')
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.savefig("RoBERTa_statement.png", format="png", dpi=400)
plt.show()
```
